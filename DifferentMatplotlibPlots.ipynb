```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.linspace(0,10,100)
```

1. Using plt.plot()

```python
plt.plot(x,y)
```

2. Using fig, ax = plt.subplots()

```python
fig,ax = plt.subplots()
ax.plot(x,y)
```

3. Adding Multiple Subplots: fig.add_subplot()

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.plot(x,y)
ax2.plot(x,y)
```

4. Using plt.subplots() for Multiple Subplots

```python
fig, axes = plt.subplots(2,2)
axes[0,0].plot(x,y)
axes[0, 1].plot(x, -y) 
axes[1, 0].plot(-x, y) 
axes[1, 1].plot(-x, -y) 
```

5. Combining Subplots with Custom Layouts

```python
figure = plt.figure(figsize=(10,6))

ax1 = figure.add_subplot(1,2,1)
ax2 = figure.add_subplot(2,2,2)
ax3 = figure.add_subplot(2,2,4)
ax1.plot(x,y)
ax2.plot(x,-y)
ax3.plot(x,y)
```
